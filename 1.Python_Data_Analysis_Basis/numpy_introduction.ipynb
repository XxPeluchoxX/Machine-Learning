# Numpy Introduction

```python
import numpy as np
np.__version__
```

## Creating ndarrays

```python
np.array([1, 4, 2, 5, 3])
```

the elements of an ndarray must be of the same type (unlike python's typical lists). If types do not match, types will be upcasted if possible

```python
np.array([3.14, 2, 4, 5])
```

```python
# Creating an array of zeros
display(np.zeros(10, dtype=int))

# Creating an array of ones
display(np.ones(10, dtype=float))
display(np.ones((2,5), dtype=float))

# Create a table full of the same value
display(np.full((4,2), 'a'))

# Create a ndarray similar to range() function
display(np.arange(0, 20, 2))

# Create a ndarray of some number of values evenly spaced
display(np.linspace(0, 1, 5))
display(np.linspace(0, 10, 6))

# Create an array of uniformly distributed random values between 0 and 1
display(np.random.random(5))
display(np.random.random((2,3)))

# Create an array of uniformly distributed random values of a
# normal with specific mean and stddev
display(np.random.normal(0, 1, 3))

# Create an array of random integers on a range [a, b)
display(np.random.randint(0, 10, 3))

# Create an identity matrix
display(np.eye(2))
```

### dtypes (data types)

At the creation of the `ndarray`s we can specify the type of its elements. We can specify the type via a string or via the numpy object

```python
display(np.zeros(10, dtype=float))
display(np.ones(10, dtype="int16"))
display(np.zeros(5, dtype=np.int16))
```

## The Basics

### Array Attributes

```python
np.random.seed(0)

x1 = np.random.randint(10, size=6)
x2 = np.random.randint(10, size=(3,4))
x3 = np.random.randint(10, size=(3,4,5))

display(x1)
display(x2)
display(x3)

print(f"The array x3 has {x3.ndim} dimensions")
print(f"The shape of each dimension is {x3.shape}")
print(f"It contains a total of {x3.size} elements")
print(f"Its elements are {x3.dtype}")
print(f"It has a size of {x3.nbytes} bytes")
```

### Array indexing

```python
display(x2)

print(x2[0, 0])
print(x2[0][0])
print(x2[2,-1])
```

### Array Slicing

`x[start:stop (not included):step]`

```python
display(x1)

display(x1[::])
display(x1[1::])
display(x1[:-1:])
display(x1[1:-1:2])
display(x1[::2])
display(x1[1::2])

print("\n")

display(x1[::-1])
```

```python
# Multidimensional arrays
display(x2)

# First two rows and 3 columns
display(x2[:2, :3])

# All rows, every odd column
display(x2[:, ::2])

display(x2[::-1, ::-1])

# Accesing the first row
display(x2[0])
display(x2[0, :])

# Accesing the first column
display(x2[:, 0])
```

*Note:* Slicing doesn't generates a copy of the `ndarray`, it generates a *view*. Therefore, if we modify a slice of the array, the original array is modified too:

```python
display(x2)

second_col = x2[:, 1]
display(second_col)

second_col[1] = 10
display(second_col)

display(x2)
```

When we want to creat an explicit copy of the data within an array or subarray, we can use the `copy` method:

```python
display(x2)
display(second_col)

second_col_cp = second_col.copy()
print("Copia realizada:")
display(second_col_cp)

second_col_cp[1] = 0
print("Modificación realizada")

display(second_col_cp)
display(second_col)
```

### Reshaping

```python
ar = np.arange(1, 10)
display(ar)

display(ar.reshape((3, 3)))
```

```python
vec = np.array([1, 2, 3])
display(vec)

column_vec = vec.reshape((3, 1))
display(column_vec)
```
