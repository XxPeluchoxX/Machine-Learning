## Creating ndarrays

```python
np.array([1, 4, 2, 5, 3])
```

the elements of an ndarray must be of the same type (unlike python's typical lists). If types do not match, types will be upcasted if possible

```python
np.array([3.14, 2, 4, 5])
```

```python
# Creating an array of zeros
display(np.zeros(10, dtype=int))

# Creating an array of ones
display(np.ones(10, dtype=float))
display(np.ones((2,5), dtype=float))

# Create a table full of the same value
display(np.full((4,2), 'a'))

# Create a ndarray similar to range() function
display(np.arange(0, 20, 2))

# Create a ndarray of some number of values evenly spaced
display(np.linspace(0, 1, 5))
display(np.linspace(0, 10, 6))

# Create an array of uniformly distributed random values between 0 and 1
display(np.random.random(5))
display(np.random.random((2,3)))

# Create an array of uniformly distributed random values of a
# normal with specific mean and stddev
display(np.random.normal(0, 1, 3))

# Create an array of random integers on a range [a, b)
display(np.random.randint(0, 10, 3))

# Create an identity matrix
display(np.eye(2))
```

### dtypes (data types)

At the creation of the `ndarray`s we can specify the type of its elements. We can specify the type via a string or via the numpy object

```python
display(np.zeros(10, dtype=float))
display(np.ones(10, dtype="int16"))
display(np.zeros(5, dtype=np.int16))
```

## The Basics

### Array Attributes

```python
np.random.seed(0)

x1 = np.random.randint(10, size=6)
x2 = np.random.randint(10, size=(3,4))
x3 = np.random.randint(10, size=(3,4,5))

display(x1)
display(x2)
display(x3)

print(f"The array x3 has {x3.ndim} dimensions")
print(f"The shape of each dimension is {x3.shape}")
print(f"It contains a total of {x3.size} elements")
print(f"Its elements are {x3.dtype}")
print(f"It has a size of {x3.nbytes} bytes")
```

### Array indexing

```python
display(x2)

print(x2[0, 0])
print(x2[0][0])
print(x2[2,-1])
```

### Array Slicing

`x[start:stop (not included):step]`

```python
display(x1)

display(x1[::])
display(x1[1::])
display(x1[:-1:])
display(x1[1:-1:2])
display(x1[::2])
display(x1[1::2])

print("\n")

display(x1[::-1])
```

```python
# Multidimensional arrays
display(x2)

# First two rows and 3 columns
display(x2[:2, :3])

# All rows, every odd column
display(x2[:, ::2])

display(x2[::-1, ::-1])

# Accesing the first row
display(x2[0])
display(x2[0, :])

# Accesing the first column
display(x2[:, 0])
```

*Note:* Slicing doesn't generates a copy of the `ndarray`, it generates a *view*. Therefore, if we modify a slice of the array, the original array is modified too:

```python
display(x2)

second_col = x2[:, 1]
display(second_col)

second_col[1] = 10
display(second_col)

display(x2)
```

When we want to creat an explicit copy of the data within an array or subarray, we can use the `copy` method:

```python
display(x2)
display(second_col)

second_col_cp = second_col.copy()
print("Copia realizada:")
display(second_col_cp)

second_col_cp[1] = 0
print("Modificación realizada")

display(second_col_cp)
display(second_col)
```

### Concatenation and Splitting

```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
display(np.concatenate([x, y]))

z = np.array([3, 3, 3])
display(np.concatenate([x, y, z]))
```

```python
# Concatenation of 2-dimensional arrays

grid = np.arange(1, 7).reshape((2, 3))
display(grid)

# Concatenate along the first axis
display(np.concatenate([grid, grid]))
display(np.concatenate([grid, grid], axis=0))

# Concatenate along the second axis
display(np.concatenate([grid, grid], axis=1))
```

```python
# We can also use vstack and hstack (vertical/horizontal) stack
display(grid)

# Vertical
last_row = np.array([9, 9, 9])
display(np.vstack([grid, last_row]))

# Horizontal
last_col = np.array([9, 9]).reshape((2, 1))
display(np.hstack([grid, last_col]))

# dstack will stack along the third axis
```

```python
# Splitting
x = np.array([1, 2, 3])
x = np.concatenate([x, z, y])
display(x)

x1, x2, x3 = np.split(x, [2, 7]) # split points
print(x1, x2, x3)

# 2-dim splitting
grid = np.arange(0, 16).reshape((4, 4))
display(grid)

upper, bottom = np.vsplit(grid, [2])
print(upper, bottom)

left, right = np.hsplit(grid, [2])
print(left, right)
```

## Universal Functions

We can perform individual operations on the whole array. It makes the syntax easier but also accelerates the code via vectorized operations.

```python
numbers = np.arange(1, 6)
display(numbers)

result = np.empty(len(numbers))
for i in range(len(numbers)):
    result[i] = 1.0 / numbers[i]
display(result)

result2 = 1.0 / numbers
display(result2)
```

```python
# We can also operate two arrays and more than one-dimensional arrays
display(np.arange(5) / np.arange(1, 6))

x = np.arange(9).reshape((3, 3))
display(2**x)
```

```python
# The operators <, <=, >, >=, ==, != are also implemented as universal functions
# They return an array of bools:
x = np.arange(0, 11)
np.random.shuffle(x)
display(x)

comp = x > 3
display(comp)

# We can also compare two different arrays
y = np.random.choice([True, False], len(x))
display(y)

z = x == y
display(z)
```

```python
# We can now count the number of elements of an array given a condition with np.count_nonzero:
display(x)
print(f'There are {np.count_nonzero(x > 3)} elements on x greater than 3')

# np.sum does the same:
print(f'There are {np.sum(x > 3)} elements on x greater than 3')

# The benefit of using np.sum is that we can choose to perform the operation on all the rows/cols at once:
x = np.random.randint(10, size=(3, 4))
display(x)

# Output the number of elements of x greater than 3 on each column and on each row
print(f'There are {np.sum(x > 3, axis=0)} elements greater than 3 on each column')
print(f'There are {np.sum(x > 3, axis=1)} elements greater than 3 on each row')
```

```python
# It is helpfull to use bitwise universal operators on boolean arrays for selecting values meeting several conditions:
# Note: It is important to use | and & rather than or and "and"
display(x)

# Counting how many elements of x are on the interval [3, 6)
print(f'There are {np.sum((x >= 3) & (x < 6))} of those elements on x')
```

## Boolean Arrays as Masks

We can use boolean arrays as masks, in order to select the part of the arrays we are interested in

```python
x = np.arange(1, 16).reshape((3,5))
display(x)

display(x > 5)

# We can form a new array with the elements we are interested in. To do that, we just need to index the array on the boolean array:
display(x[x > 5])
```

## Fancy Indexing

```python
x = np.random.randint(100, size=10)
display(x)

# Instead of doing this to work with the desired elements:
x_desired = np.array([x[1], x[3], x[7]])
display(x_desired)

# We can use fancy indexing and index the array on a list of indexes:
display(x[[1, 3, 7]])

# The indexed array reflexes the index array and not the original one. For example:
indexes = np.array([[1, 3], [7, 8]])
display(x[indexes])
```

```python
# We can also do fancy indexing in multiple-dimension arrays:
x = np.arange(1, 16).reshape((3, 5))
display(x)

rows = [0, 1, 2]
cols = rows

display(x[rows, cols])
```
