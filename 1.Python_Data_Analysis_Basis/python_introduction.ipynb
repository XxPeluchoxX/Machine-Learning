# 1.- CardDeck example

````python
# Exameple - A pythonic card deck
import collections

Card = collections.namedtuple('Card', ['rank', 'suit'])

class FrenchDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')
    suits = 'espadas diamantes picas corazones'.split()

    def __init__(self):
        self._cards = [Card(rank, suit) for rank in self.ranks for suit in self.suits]

    def __len__(self):
        return len(self._cards)

    def __getitem__(self, pos):
        return self._cards[pos]
````

````python
deck = FrenchDeck()
display(len(deck))
display(deck[0])

# Because we've implemented the __getitem__ method we have available:
import random
display(random.choice(deck))

# We also have slicing
display(deck[1:3])

# __getitem__ allow us to check if the container have an item (sequentially)
display(Card('2', 'diamantes') in deck)
display(Card('12', 'diamantes') in deck)

# And we can also iterate them
for card in deck:
    print(card)
````

# 2.- Vector Example

````python
import math

class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __repr__(self):
        return f'Vector({self.x!r}, {self.y!r})'

    def __abs__(self):
        return math.hypot(self.x, self.y)

    def __bool__(self):
        return bool(abs(self))

    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Vector(x, y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)
````

````python
v1 = Vector(1, 2)
display(v1)

v2 = Vector(3, 4)
display(v2)

display(v1 + v2)

display(v2 * 3)

display(abs(v1))
````

1. Given a list of integers, return a new list with the square of the even numbers

```
[1, 2, 3, 4, 5] -> [4, 16]
```

````python
def ex1(numbers):
    return [n*n for n in filter(lambda n: n % 2 == 0, numbers)]
````

````python
numbers = [i for i in range(1, 31)]
result = ex1(numbers)
result
````

3. Group people by city:
```
[(Ana, Granada), (Luis, Madrid), (Eva, Granada)] -> [(Granada, (Ana, Eva)), (Madrid, (Luis,))]
```

````python
people = [
    ('Ana', 'Granada'),
    ('Marta', 'Segovia'),
    ('Luisa', 'Cáceres'),
    ('Martín', 'Badajoz'),
    ('Matías', 'Granada'),
    ('Julia', 'Cáceres'),
    ('Rubén', 'Pontevedra'),
    ('Mikael', 'Cáceres'),
]

from collections import defaultdict

ne = defaultdict(list)
for name, city in people:
    ne[city].append(name)

for city in ne:
    print(f'{city}: ')
    for neigh in ne[city]:
        print(f'\t{neigh}')
````
