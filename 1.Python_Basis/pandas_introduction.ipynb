# Pandas Introduction

```python
import numpy as np
import pandas as pd
```

## Objects

### Series

```python
# Pandas Series
data = pd.Series([0.25, 0.5, 0.75, 1.0])
display(data)

display(data.values) # numpy array
display(data.index)
```

```python
# Is somehow a Python dictionary, but it is more efficient
data = pd.Series([0.25, 0.5, 0.75, 1.0], index=['a', 'b', 'c', 'd'])
display(data)

data = pd.Series([0.25, 0.5, 0.75, 1.0], index=[2, 5, 3, 7])
display(data)
```

```python
population_dict = {
    'California': 123456789,
    'Texas': 97654321,
    'New York': 12341234124,
    'Florida': 987987987,
}
population = pd.Series(population_dict)
display(population)

display(population['California':'New York'])
```

### DataFrame

```python
# We can combine two Series object to make a DataFrame object
display(population)

area = pd.Series({
    'California': 123123,
    'Texas': 98098,
    'Florida': 123456,
    'New York': 98765,
})
display(area)

states = pd.DataFrame({'population': population, 'area': area})
display(states)

display(states.index)
display(states.columns)
```

```python
# We can think about a DataFrame like a dictionary which maps column name to a Series of column data
display(states['area'])
display(states['population'])
```

#### Construction of DataFrame objects

```python
# From a series object
x = pd.DataFrame(population, columns=['population'])
display(x)

# From a list of dicts
x = pd.DataFrame([{'a': i, 'b': 2 * i} for i in range(3)])
display(x)

# If a value is missing pandas will fill up its values with NaN
x = pd.DataFrame([{'a': 1, 'b': 2}, {'a': 3, 'c': 14}])
display(x)
```

```python
# From a dictionary of Series objects
x = pd.DataFrame({'population': population, 'area': area})
display(x)

# From a two-dimensional Numpy array
x = pd.DataFrame(
    np.random.rand(3, 2),
    columns=['foo', 'bar'],
    index=['a','b','c'],
)
display(x)
```

### Index Object
Can be seen as an inmutable array or a multiset (thus its supports intersection and union operations)

```python
ind = pd.Index([2, 3, 5, 7, 11])
display(ind)

# Operates like an array
display(ind[1])
display(ind[0:2])
print(ind.size, ind.shape, ind.ndim, ind.dtype)

# They are inmutable
ind[0] = 4
```

## Indexing and Selection

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index=['a', 'b', 'c', 'd'])
display(data)

display(data['b'])

# Like Python dictionaries
display(data.keys())
display(list(data.items()))
data['e'] = 0.8

# Fancy indexing
display(data[['a', 'e']])

# masking
display(data[(data > 0.3) & (data < 0.8)])
```

```python
data = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
display(data)

# Explicit index at indexing
display(data[1])

# Implicit index at slicing
display(data[1:3])
```

```python
# We shall use `loc` and `iloc` methods:
# loc usses the explicit indexes
display(data.loc[1])
display(data.loc[1:3])

# iloc uses the implicit indexes
display(data.iloc[1])
display(data.iloc[1:3])
```

```python
data = pd.DataFrame({'area': area, 'pop': population})
display(data)

# We can add a new column
data['density'] = data['pop'] / data['area']
display(data)

# We can access the values of the DataFrame as a two-dimensional numpy array
display(data.values)
```

```python
# loc and iloc can be used too
display(data.loc[:'New York', :'pop'])
display(data.iloc[:3, :2])
```

## Operating on Data
Any numpy's ufunc will work on a pandas Series or DataFrame object

```python
# When some data is NaN and we do an operation it will be NaN:
area2 = pd.Series({'Alaska': 1234, 'Texas': 9866, 'New York': 4567})
pop2 = pd.Series({'California': 987, 'Texas': 123, 'New York': 444})
display(area2 / pop2)

# Similarly:
A = pd.Series([2, 4, 6], index=[0, 1, 2])
B = pd.Series([1, 3, 5], index=[1, 2, 3])
display(A + B)
```

```python
# We can specify the parameter fill_value to fill out all the NaNs values:
display(A.add(B, fill_value=0))
```

## Handling Missing Data
Pandas has `None` and `NaN` as special centinels. The first one is an object and the second one is a floating numeric value

```python
display(data.isnull()) # Boolean mask indicating missing values
display(data.notnull()) # Opposite of the above
display(data.dropna()) # Filtered version of the data
display(data.fillna(0)) # Returns a copy with missing values inmputed
```

## Aggregation

```python
# We'll work with the planets dataset
import seaborn as sns
planets = sns.load_dataset('planets')
display(planets.shape)
display(planets.head())
```

In pandas the aggregates returns by default the result on each column

```python
planets_num = planets.select_dtypes(include='number')
display(planets_num.head())

display(planets_num.mean())

# We can use the axis argument to change its behaviour
display(planets_num.mean(axis='columns').head())
```

```python
# 5.- Select all the planets discovered after 2010
display(planets[planets['year'] >= 2010])

# 6.- Select all the planets discovered after 2010 with a non-NaN mass with a distance greater than 100
display(planets.dropna()[(planets['year'] >= 2010) & (planets['distance'] > 100)])
```

```python
# 7.- Count the number of null values on a column
planets.isnull().sum()
```

```python
# 8.- Remove the rows with some null value
display(planets.dropna().head())

# 9.- Fill up the null values of distance column by its mean
planets['distance'] = planets['distance'].fillna(planets['distance'].mean())
display(planets.head())
```

```python
# 10.- Get all categories from the method's column
display(pd.unique(planets['method']))

# 11.- Generate a new column
planets['O/D'] = planets['orbital_period'] / planets['distance']
display(planets.head())
```

```python
# 12.- Rename some column
planets = planets.rename(columns={'O/D': 'Orb/Dist'})
display(planets.head())

# 13.- Drop a column
planets = planets.drop(columns='Orb/Dist')
display(planets.head())
```

```python
# 14.- Sorts the dataframe by the year
planets = planets.sort_values(by='year')
display(planets.head())
```

```python
# 15.- Transform the column year xxxx into 'x.xxx'
planets['year'] = planets['year'].map(lambda x: f'{x:,}'.replace(',', '.'))
display(planets.head())
```

```python
# 16.- Compute the average distance by year
display(planets.groupby('year')['distance'].mean())
```

```python
# 17.- For each method compute: the maximum distance, the number of discovers and the medium mass
planets['distance'] = planets['distance'].fillna(planets['distance'].mean())
planets['mass'] = planets['mass'].fillna(planets['mass'].mean())
display(planets.groupby('method').agg({
    'distance': 'max',
    'method': 'count',
    'mass': 'mean',
}))
```

```python
# 18.- Build two DataFrames: The df without the mass and a numpy's array with the mass
y = planets['mass']
display(y.head())

X = planets.drop(columns='mass')
display(X.head())

display(planets.head())

# 19.- Transform year into numeric and remove from X the non-numeric columns
X['year'] = X['year'].map(lambda x: int(x[0] + x[2:]))
X = X.select_dtypes(include='number')
display(X.head())
```

```python
# 20.- One-Hot Encode a column 
display(X.head())

X = pd.get_dummies(X, columns=['number'])
display(X.head())
```

```python
# 21.- Show and remove the duplicated rows
dupl = planets[planets.duplicated(keep=False)]
print(f'There are {len(dupl)} duplicated rows:')
display(dupl)

display(planets.shape)
planets = planets.drop_duplicates()
print('Duplicated lines removed')
display(planets.shape)
```

```python
# 22.- Return the 10th more furthest planets
display(planets.nlargest(10, 'distance'))
```

```python
# 23.- Change the dtype of a column
display(planets.dtypes)

planets['year'] = planets['year'].astype(float)
display(planets.head())
display(planets.dtypes)
```
